```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#preprocessing the input data
data = pd.read_csv('data.csv')
X = data.iloc[:,0]
Y = data.iloc[:,1]
plt.scatter(X,Y)
plt.show()

```

```python
#Building the model
m = 0
c = 0

L = 0.0001
iterations = 1000

n = float(len(X))

for i in range(len(X)):
    Y_pred = m*X+c
    D_m = (-2/n)*sum(X * (Y - Y_pred))
    D_c = (-2/n)*sum(Y - Y_pred)
    m = m - L*D_m
    c = c - L*D_c
    
print (m,c)
```

```python
#Making predictions
Y_pred = m*X+c

plt.scatter(X,Y)
plt.plot([min(X),max(X)],[min(Y_pred), max(Y_pred)],color='red')
plt.show()
```
